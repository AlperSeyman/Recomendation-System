```python
import numpy as np
import pandas as pd

# data set link : "https://www.kaggle.com/datasets/grouplens/movielens-20m-dataset"
```

```python
movie = pd.read_csv("movie.csv")
movie.head()
```

```python
movie.info()
```

```python
movie = movie.iloc[:,[0,1]]
movie.head(10)
```

```python
rating = pd.read_csv("rating.csv")
rating.head()
```

```python
rating = rating.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
data = pd.merge(movie,rating)
data.head(5)
```

```python
data = data.iloc[:1000000,:]
data.head()
```

```python
pivot_table = data.pivot_table(index = ["userId"], columns=["title"], values="rating")
pd.set_option('display.max_columns', None)
pivot_table.head(10)
```

```python
movie_watched = pivot_table["Angels and Insects (1995)"]

# Similarty with other movies
similarity_rate = pivot_table.corrwith(movie_watched)
similarity_rate = similarity_rate.sort_values(ascending=False)
similarity_rate.head(5)
```
